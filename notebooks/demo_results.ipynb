## 3. Test `hello_world()`

```python
# Prints and returns the greeting
result = autoseachlib.hello_world()
print(f"Return value: {result!r}")
```

## 4. Test `add_strings()`

```python
# Concatenate two strings
result = autoseachlib.add_strings("Hello, ", "World!")
print(f"Result: {result!r}")
```

```python
# Try with custom strings
name = autoseachlib.add_strings("CarScan", ".ai")
greeting = autoseachlib.add_strings("Welcome to ", name)
print(greeting)
```

```python
# Error handling — passing a non-string raises TypeError
try:
    autoseachlib.add_strings(42, "oops")
except TypeError as e:
    print(f"Caught expected error: {e}")
```
